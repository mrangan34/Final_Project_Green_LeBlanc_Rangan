```python
# Import dependencies
import numpy as np
import pandas as pd
```

```python
# Import music datasets
#csv file
song_data1 = pd.read_csv('song_data.csv')
print(song_data1)
```

```python
#txt file
song_data2 = pd.read_fwf('songdata_10000.txt')
song_data2.columns = ['user_id','song_id','listen_count']
print(song_data2)
```

```python
# merge datasets and remove duplicates
song_df = pd.merge(song_data2, song_data1.drop_duplicates(['song_id']), on= "song_id", how='left')
song_df
```

```python
#write file in csv to be visualized in Tableau
song_df.to_csv('final_song_df.csv', index=False)
```

```python
# Simialar Tracks Data Wrangle
song_grouped = pd.DataFrame(song_df.groupby('song_id')['listen_count'].count())
song_grouped
```

```python
song_grouped.dtypes
```

```python
song_df.dtypes
```

```python
#Changing columns data types (song_id cast to string so that it can be queried)
convert_song_df = song_df.astype({'listen_count':'int32'})
```

```python
convert_song_df.dtypes
```

```python
convert_song_df['song_id'] = convert_song_df['song_id'].astype('str')
```

```python
convert_song_df.dtypes
```

```python
#Query dataframe for specific song id (Nickelback: Far Away)
convert_song_df[convert_song_df['song_id'] == 'SOFVZRE12A8C139783']
```

```python
# Pivot table that shows the number of times each user listened to each of the songs in the dataset
songs_crosstab = pd.pivot_table(convert_song_df, values = 'listen_count', index = 'user_id', columns = 'song_id')
songs_crosstab.head()
```

```python
# Predictor Song Ratings -- how many times each user listen to 'Far Away' by Nickelback
predictor_song_ratings = songs_crosstab['SOFVZRE12A8C139783']
predictor_song_ratings
```

```python
# List of users who listened to 'Far Away' by Nickelback more than 1 time
predictor_song_ratings[predictor_song_ratings>= 1]
```

```python
# Find correlation between 'Far Away' and other songs in the dataset
similar_songs = songs_crosstab.corrwith(predictor_song_ratings)
```

```python
# Create new dataframe with correlated values of songs to 'Far Away'
corr_listened_song = pd.DataFrame(similar_songs, columns = ['pearsonR'])
corr_listened_song.dropna(inplace = True)
```

```python
corr_listened_song
```

```python
# Join the correlation dataframe with the grouped songs dataframe
# to see the correlation value next to each song
predictor_corr_summary = corr_listened_song.join(song_grouped['listen_count'])
```

```python
# place higher correlated songs at the top of the dataframe
predictor_corr_summary = predictor_corr_summary.sort_values('pearsonR', ascending = False)
predictor_corr_summary
```

```python
# final recommendation
final_recommended_songs = predictor_corr_summary[predictor_corr_summary.pearsonR < 0.9999]
final_recommended_songs.sort_values('pearsonR', ascending = False)
final_recommended_songs = final_recommended_songs.reset_index()
final_recommended_songs
```

```python
song_df_one = song_df.drop(['listen_count'], axis=1)
song_df_one
```

```python
similar_songs = pd.merge(final_recommended_songs, song_df_one.drop_duplicates(["song_id"]), on="song_id", how="left")
similar_songs = similar_songs.sort_values('pearsonR', ascending = False)
similar_songs.head(50)
```
