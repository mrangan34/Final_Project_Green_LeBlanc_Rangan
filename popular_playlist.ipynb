```python
import pandas as pd
```

```python
# Import music datasets
#csv file
song_data1 = pd.read_csv('song_data.csv')

#txt file
song_data2 = pd.read_fwf('songdata_10000.txt')
song_data2.columns = ['user_id','song_id','listen_count']
```

```python
song_data = pd.DataFrame(song_data1)
othersong_data = song_data2
```

```python
song_df = pd.merge(othersong_data, song_data.drop_duplicates(['song_id']), on='song_id', how ='left')
song_grouped = song_df.groupby(['title']).agg({'listen_count':'count'})
grouped_sum = song_grouped['listen_count'].sum()
```

```python
# Calculate percent share of each song in listen count
song_grouped['percentage'] = song_grouped['listen_count'].div(grouped_sum)*100
```

```python
# Sort the dataset by listen count
song_grouped = song_grouped.sort_values(['listen_count'], ascending = True)
song_df = song_df['listen_count'].astype(float)
popular = song_grouped.sort_values(by = 'listen_count')
```
